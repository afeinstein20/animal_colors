```python
import numpy as np
from animal_colors import *
import matplotlib.pyplot as plt
```

```python
sens = Sensitivity('house fly')
```

```python
sens.plot()
```

```python
anim_cm = AnimalColormaps(sens)
```

```python
anim_cm.plot_colorwheel()
```

```python
anim_cm.alter_hsv(saturation=0.6, value=1.0, hue=0.9)

anim_cm.plot_colorwheel(colors=anim_cm.new_rgb)
```

```python
c,_ = anim_cm.tetradic(80)
c1,_= anim_cm.tetradic(80, colors=anim_cm.new_rgb)
for i,color in enumerate(c):
    plt.plot(0.01, i/2.0, 'o', c=color, ms=15)
    plt.plot(0.02, i/2.0, 'o', c=c1[i], ms=15)
plt.xlim(0,0.03)
```

```python
from matplotlib.colors import LinearSegmentedColormap

clist =  np.flip(c1)
#clist = np.append(clist, '#000000')

cm = LinearSegmentedColormap.from_list('sequential', 
                                       c1,
                                       N=2048)

plt.imshow(np.full((100,1000),np.arange(0,1000,1)), cmap=cm)
```
