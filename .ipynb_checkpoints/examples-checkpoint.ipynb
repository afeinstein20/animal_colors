```python
import numpy as np
from animal_colors import *
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 16
```

```python
sens = Sensitivity('house fly')
fly_map = sens.mapped+0.0

sens = Sensitivity('human')
human_map = sens.mapped+0.0

sens = Sensitivity('blue tit')
tit_map = sens.mapped+0.0
```

```python
lw=4
human = '#b5b5b5'
fly = '#5c5b5b'
tit = '#000000'
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True, sharey=True,
                               figsize=(10,6), facecolor='w')

ax1.fill_between(sens.wave_x, human_map[:,0], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, label='Human', alpha=0.6)
ax1.fill_between(sens.wave_x, human_map[:,1], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, alpha=0.6)
ax1.fill_between(sens.wave_x, human_map[:,2], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, alpha=0.6)


ax1.fill_between(sens.wave_x, fly_map[:,0], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=fly, label='House Fly', alpha=0.7)
ax1.fill_between(sens.wave_x, fly_map[:,1], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=fly, alpha=0.7)
ax1.fill_between(sens.wave_x, fly_map[:,2], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=fly, alpha=0.7)

ax1.plot(sens.wave_x, fly_map[:,0], 
                 lw=lw, color=fly)
ax1.plot(sens.wave_x, fly_map[:,1], 
                 lw=lw, color=fly)
ax1.plot(sens.wave_x, fly_map[:,2],
                 lw=lw, color=fly)


ax1.fill_between(sens.wave_x+10000, fly_map[:,2], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=tit, alpha=0.7, label='Blue Tit')


ax2.fill_between(sens.wave_x, human_map[:,0], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, label='Human', alpha=0.6)
ax2.fill_between(sens.wave_x, human_map[:,1], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, alpha=0.6)
ax2.fill_between(sens.wave_x, human_map[:,2], np.zeros(len(human_map[:,0])), 
                 lw=lw, color=human, alpha=0.6)

ax2.fill_between(sens.wave_x, tit_map[:,0], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=tit, label='Blue Tit', alpha=0.7)
ax2.fill_between(sens.wave_x, tit_map[:,1], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=tit, alpha=0.7)
ax2.fill_between(sens.wave_x, tit_map[:,2], np.zeros(len(fly_map[:,0])), 
                 lw=lw, color=tit, alpha=0.7)

ax2.plot(sens.wave_x, tit_map[:,0], 
                 lw=lw, color=tit)
ax2.plot(sens.wave_x, tit_map[:,1], 
                 lw=lw, color=tit)
ax2.plot(sens.wave_x, tit_map[:,2],
                 lw=lw, color=tit)


ax1.legend(bbox_to_anchor=(0., 1.02, 1., .102),
                   loc='lower left',
                   ncol=sens.mapped.shape[1], mode="expand",
                   borderaxespad=0.)
ax2.set_xlabel('Wavelength [nm]')
ax2.set_ylabel('Normalized Spectral Sensitivities', y=1.01)

plt.xlim(300,700)
plt.ylim(0,1.03)
plt.savefig('./plots/comparison.pdf', rasterize=True, bbox_inches='tight',
            dpi=300)
```

```python
anim_cm = AnimalColormaps(sens)
```

```python
anim_cm.plot_colorwheel()
```

```python
anim_cm.alter_hsv(saturation=0.6, value=1.0, hue=0.9)

anim_cm.plot_colorwheel(colors=anim_cm.new_rgb)
```

```python
c,_ = anim_cm.tetradic(80)
c1,_= anim_cm.tetradic(80, colors=anim_cm.new_rgb)
for i,color in enumerate(c):
    plt.plot(0.01, i/2.0, 'o', c=color, ms=15)
    plt.plot(0.02, i/2.0, 'o', c=c1[i], ms=15)
plt.xlim(0,0.03)
```

```python
cm = anim_cm.make_colormap(hex_values=c1)

plt.imshow(np.full((100,1000), np.linspace(0,1,1000)), cmap=cm)
```
